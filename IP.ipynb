```python
df_numerico['FUNDAMENTAL_PAI'] = np.where(
    df_numerico['ESCOLARIDADE_PAI'] == "ENSINO FUNDAMENTAL INCOMPLETO", 1, np.where(
    df_numerico['ESCOLARIDADE_PAI'] !=  "ENSINO FUNDAMENTAL INCOMPLETO", 0, -1))
df_numerico['ENSINO MEDIO_PAI'] = np.where(
    df_numerico['ESCOLARIDADE_PAI'] == "ENSINO MEDIO", 1, np.where(
    df_numerico['ESCOLARIDADE_PAI'] !=  "ENSINO MEDIO", 0, -1))
df_numerico['ENSINO SUPERIOR_PAI'] = np.where(
    df_numerico['ESCOLARIDADE_PAI'] == "ENSINO SUPERIOR", 1, np.where(
    df_numerico['ESCOLARIDADE_PAI'] !=  "ENSINO SUPERIOR", 0, -1))

df_numerico['FUNDAMENTAL_MAE'] = np.where(
    df_numerico['ESCOLARIDADE_MAE'] == "ENSINO FUNDAMENTAL INCOMPLETO", 1, np.where(
    df_numerico['ESCOLARIDADE_MAE'] !=  "ENSINO FUNDAMENTAL INCOMPLETO", 0, -1))
df_numerico['ENSINO MEDIO_MAE'] = np.where(
    df_numerico['ESCOLARIDADE_MAE'] == "ENSINO MEDIO", 1, np.where(
    df_numerico['ESCOLARIDADE_MAE'] !=  "ENSINO MEDIO", 0, -1))
df_numerico['ENSINO SUPERIOR_MAE'] = np.where(
    df_numerico['ESCOLARIDADE_MAE'] == "ENSINO SUPERIOR", 1, np.where(
    df_numerico['ESCOLARIDADE_MAE'] !=  "ENSINO SUPERIOR", 0, -1))

df_numerico.drop('ESCOLARIDADE_PAI', axis=1, inplace=True)
df_numerico.drop('ESCOLARIDADE_MAE', axis=1, inplace=True)

df_numerico
```
